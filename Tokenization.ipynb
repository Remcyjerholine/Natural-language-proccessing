```python
import nltk  #natural language  toolkit 
nltk.download('punkt')  #punkt is an tokenization model and Instead of relying on hardcoded rules..it had other understanding of the starting and ending context
nltk.download('punkt_tab') # Required in newer NLTK versions
from nltk.util import ngrams   #used to group the words into tuple
from nltk.tokenize import word_tokenize  #used to split the word from the sentence
text = 'this is my cat this is my dog'  #the input sentence
tokens = word_tokenize(text.lower())    #creating an variable and doing tokenization after case folding
print("Unigrams:", list(ngrams(tokens, 1)))  #one in an tuple
print("Bigrams:", list(ngrams(tokens, 2)))   #two in a tuple
print("Trigrams:", list(ngrams(tokens, 3)))  #three in a tuple
```
