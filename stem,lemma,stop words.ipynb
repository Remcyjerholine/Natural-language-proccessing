```python
import nltk #natural language toolkit
# Download required resources
nltk.download('punkt')  #downloading the punkt which is an tikenization model 
nltk.download('punkt_tab') #new lib of punkt
nltk.download('stopwords') #list of stop words
nltk.download('wordnet')   #contain all english library words for lemma and stem
nltk.download('omw-1.4')   #downloads all the Open Multilingual WordNet
from nltk.tokenize import word_tokenize, sent_tokenize #import for tokenization
from nltk.corpus import stopwords                      #import all the stopwords list which is downloaded
from nltk.stem import PorterStemmer, WordNetLemmatizer #import for stem and lemma
# Input Text
text = """
Natural Language Processing is a branch of Artificial Intelligence.
It enables computers to understand human language and process text efficiently.
""" # three qoutes for entering multiple lines
# Sentence Tokenization
sentences = sent_tokenize(text) #creating an variable which stores the breaked sentences from the paragraph
# Word Tokenization
words = word_tokenize(text)v #creating an variable which contain all the tokens from the sentences 
# Stopword Removal
stop_words = set(stopwords.words('english')) #store all the stop words in the variable created
filtered_words = [word for word in words if word.lower() not in stop_words] #converting all the text in lower and then if its not it the stop word list ,keep it
# Stemming
stemmer = PorterStemmer()  #storing all the functions of the stemming function in the variable created
stemmed_words = [stemmer.stem(word) for word in filtered_words] 
# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_words = [lemmatizer.lemmatize(word) for word in filtered_words]
# Display Results
print("Sentence Tokens:")
print(sentences)
print("\nWord Tokens:")
print(words)
print("\nWords After Stopword Removal:")
print(filtered_words)
print("\nStemmed Words:")
print(stemmed_words)
print("\nLemmatized Words:")
print(lemmatized_words)
```
